# Check Existing Python Environment

```python
# Check for existing environment files
import os
env_files = ['pyproject.toml', 'requirements.txt', 'Pipfile', 'poetry.lock', 'uv.lock']
for file in env_files:
    if os.path.exists(file):
        print(f"Found: {file}")
    else:
        print(f"Not found: {file}")

# Check if .venv exists
if os.path.exists('.venv'):
    print(".venv directory exists")
else:
    print(".venv directory does not exist")

# Current Python
import sys
print(f"Python executable: {sys.executable}")
print(f"Python version: {sys.version}")
```

# Set Up Environment with uv and .venv

```python
# To set up the environment:
# 1. Install uv if not installed
# 2. Run: uv venv .venv
# 3. Run: uv sync
# This creates a local .venv and installs dependencies from pyproject.toml

print("Environment setup commands:")
print("uv venv .venv")
print("uv sync")
```

# Load and Inspect Dataset

```python
import pandas as pd

# Load the dataset
file_path = '../data/raw/hdb_presale_prices_2015-2024_cleaned_regression.csv'
df = pd.read_csv(file_path)

# Basic inspection
print("Dataset shape:", df.shape)
print("\nColumns:", list(df.columns))
print("\nData types:")
print(df.dtypes)
```

```python
print("\nFirst 5 rows:")
print(df.head())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
print("\nUnique values in categorical columns:")
categorical_cols = ['town', 'flat_type', 'storey_range', 'storey_range_category', 'distance_from_expressway']
for col in categorical_cols:
    unique_vals = df[col].unique()
    print(f"{col}: {len(unique_vals)} unique values - {unique_vals[:5]}...")  # Show first 5
```
